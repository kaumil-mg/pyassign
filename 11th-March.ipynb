Q1: The main difference between a t-test and a z-test lies in the population standard deviation and the sample size. A t-test is used when the population standard deviation is unknown and/or the sample size is small, while a z-test is used when the population standard deviation is known and the sample size is large.

Example scenario for a t-test: A university wants to compare the effectiveness of two different teaching methods on student performance. They conduct a study with two groups of students, each taught using a different method, and compare their exam scores. Since the population standard deviation of exam scores is unknown, and the sample size is relatively small, a t-test would be appropriate to determine if there is a significant difference in the mean exam scores between the two groups.

Example scenario for a z-test: A pharmaceutical company wants to test whether a new drug reduces blood pressure. They have historical data showing the population standard deviation of blood pressure reductions for similar drugs. They conduct a clinical trial with a large sample size and find that the mean reduction in blood pressure is significantly different from zero. In this case, a z-test would be appropriate due to the known population standard deviation and large sample size.

Q2: One-tailed tests and two-tailed tests differ in their hypotheses and the directionality of the significance testing. 

- One-tailed tests: These tests focus on a specific direction of the relationship or difference. They are used when there is a specific hypothesis about the direction of the effect. For example, testing whether a new drug increases reaction time.
  
- Two-tailed tests: These tests examine the possibility of a relationship or difference in both directions. They are used when there is no specific hypothesis about the direction of the effect. For example, testing whether a new teaching method has any effect on exam scores.

Q3: Type 1 and Type 2 errors are potential errors in hypothesis testing.

- Type 1 error: It occurs when the null hypothesis is wrongly rejected when it is actually true. This is also known as a false positive. For example, rejecting the null hypothesis that a drug has no effect when it actually doesn't.

- Type 2 error: It occurs when the null hypothesis is wrongly accepted when it is actually false. This is also known as a false negative. For example, accepting the null hypothesis that a drug has no effect when it actually does.

Q4: Bayes's theorem is a mathematical formula used to calculate the conditional probability of an event occurring, given prior knowledge and new evidence. It is expressed as:

\[ P(A|B) = \frac{{P(B|A) \times P(A)}}{{P(B)}} \]

Where:
- \( P(A|B) \) is the probability of event A occurring given that event B has occurred.
- \( P(B|A) \) is the probability of event B occurring given that event A has occurred.
- \( P(A) \) is the prior probability of event A occurring.
- \( P(B) \) is the prior probability of event B occurring.

Example: Suppose a factory produces two types of products, A and B. Product A has a defect rate of 10%, while product B has a defect rate of 5%. If a randomly selected product is defective, Bayes's theorem can be used to calculate the probability that it is product A.

Q5: A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence. It provides an estimate of the precision of the sample statistic.

To calculate a confidence interval, you need:
- Sample mean (\( \bar{x} \))
- Sample standard deviation (\( s \))
- Sample size (\( n \))
- Confidence level (typically expressed as a percentage, such as 95% or 99%)

Example: Let's say we want to calculate the 95% confidence interval for the average height of students in a school. We measure the heights of 100 randomly selected students and find that the sample mean height is 65 inches with a sample standard deviation of 3 inches. Using the t-distribution for a 95% confidence level and a sample size of 100, we calculate the confidence interval as \( \bar{x} \pm t \times \frac{s}{\sqrt{n}} \).
