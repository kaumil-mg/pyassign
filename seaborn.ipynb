Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a data visualization library built on top of Matplotlib. Some of the plots that can be created using Seaborn library and their uses are:

- Scatter Plot - It is used to represent the relationship between two variables. Seaborn’s scatterplot() function also provides options to add a third variable to represent the data points, which can be represented by the size, color, and shape of the markers.

- Line Plot - It is used to represent the trend of one or more variables with respect to another variable. Seaborn’s lineplot() function also provides the option to add hue and style parameters for grouping the data based on a categorical variable.

- Bar Plot - It is used to represent categorical data with rectangular bars with heights or lengths proportional to the values that they represent. Seaborn’s barplot() function can be used to display the relationship between a categorical variable and a continuous variable.

- Histogram - It is used to represent the distribution of a single numerical variable. Seaborn’s histplot() function provides the option to adjust the number of bins, control the density of the plot, and to display a kernel density estimate of the data.

- Heatmap - It is used to represent the correlation between two or more variables using a color-coded matrix. Seaborn’s heatmap() function is useful in visualizing the correlation matrix between different variables.

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.

import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset('fmri')

# Plot the lineplot using different events and regions
sns.lineplot(x='timepoint', y='signal', hue='event', style='region', data=fmri)
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

# Load the dataset
titanic_data = sns.load_dataset("titanic")

# Plot the box plots
sns.boxplot(x="pclass", y="age", data=titanic_data)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

diamonds = sns.load_dataset('diamonds')

sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.

iris = sns.load_dataset('iris')
sns.pairplot(data=iris, hue= 'species')
```
